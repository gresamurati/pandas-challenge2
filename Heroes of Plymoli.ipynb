```python
import pandas as pd
import numpy as np
```

```python
file_to_load ="purchase_data.csv"
```

```python
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

```python
#display total number of players
total_players=purchase_data["SN"].nunique()
total_players
```

```python
total_df= pd.DataFrame({'Total Players':[total_players]})
total_df
```

```python
#Number of Unique Items
unique_items=purchase_data["Item ID"].value_counts(dropna=False).count()
unique_items
```

```python
unique_df=unique_items
```

```python
#purchase count
purchase_number=purchase_data["Purchase ID"].count()
purchase_number
```

```python
purchase_df=purchase_number
```

```python
#total price
total_price=purchase_data["Price"].sum()
total_price
```

```python
total_df=total_price
```

```python
#average price
average_price=purchase_data["Price"].mean()
average_price
```

```python
average_df=average_price
```

```python
# Purchasing Analysis (Total)
purchasing_analysis = pd.DataFrame({'Number of Unique Items': [unique_df], 'Average Price': [average_df],'Number of Purchases': [purchase_df], 'Total Revenue': [total_df]}, index=[0])
purchasing_analysis 
```

```python
# Gender Demographics
#count of females
females=len(purchase_data[purchase_data["Gender"]=="Female"]['SN'].unique())
females
#count of other
other=len(purchase_data[purchase_data["Gender"]=="Other / Non-Disclosed"]['SN'].unique())
#Count of male Players
males=len(purchase_data[purchase_data["Gender"]=="Male"]['SN'].unique())
males
#female percentage
female_percentage=round(females/total_players*100, 2)
female_percentage
#male percentage
male_percentage=round(males/total_players*100, 2)
male_percentage
#Percentage of Other / Non-Disclosed
other_percentage=round(other/total_players*100, 2)
other_percentage
```

```python
gender_data={'Total Count':[males, females, other], 'Percentage of Players':[male_percentage, female_percentage,other_percentage]}
```

```python
purchase_count = purchase_data.groupby('Gender').count()
purchase_count=purchase_count.iloc[:,:1].reset_index()
purchase_count.columns = ['Gender','Purchase Count']
purchase_count
```

```python
#setting the index by unique values
gender_demographics = pd.DataFrame(gender_data, index =['Male', 'Female', 'Other / Non-Disclosed']) 
gender_demographics['Percentage of Players']=gender_demographics['Percentage of Players'].astype(float).map(
    "%{:,.2f}".format)
gender_demographics
```

```python
#getting the mean of price
purchase_mean=purchase_data.groupby('Gender').mean()
purchase_mean=purchase_mean.iloc[:,3:4].reset_index()
purchase_mean.columns = ['Gender', 'Average Purchase Price']
purchase_mean.round(2)
purchase_mean
```

```python
#merging the above
final_df = purchase_count.merge(purchase_mean)
final_df.round(2)
```

```python
#getting the total purchase value
total_value=purchase_data.groupby('Gender').sum()
total_value=total_value.iloc[:,3:4].reset_index()
total_value.columns=['Gender','Total Purchase Value']
total_value
```

```python
#merging
final_df=final_df.merge(total_value)
final_df
```

```python
final_df['Totalcounts']=[females,males,other]
```

```python
#calculating avg total purchase per person
final_df['Avg Total Purchase per Person']  = final_df['Total Purchase Value'] / final_df['Totalcounts']
```

```python
#rounding
final_df = final_df.round(2)
```

```python
#removing totalcounts
final_df=final_df.drop(columns=['Totalcounts'])
final_df
```

```python
#setting index to gender values
final_df=final_df.set_index('Gender')
final_df
```

```python
#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
#setting the bins
bins=[0, 9, 14, 19, 24, 29, 34, 39, 99]

#setting the column to  use for the bins
Age_groups=pd.cut(purchase_data['Age'], bins, labels=['<10', '10-14', '15-19', '20-24', '25,29', '30-34', '35-39', '40+'])
print(Age_groups)
```

```python
purchase_data
```

```python
purchase_data['Age Ranges']=Age_groups
```

```python
#getting the unique values in the age column and renaming the bins
unique_SN=purchase_data.groupby('Age Ranges').nunique()
unique_SN=unique_SN.iloc[:,1:2].reset_index()
unique_SN.columns=['Age Ranges','Unique people']
```

```python
unique_SN
```

```python
#Average Purchase Price
mean_data=purchase_data.groupby('Age Ranges').mean()

mean_data=mean_data.iloc[:, 3:4].reset_index()

mean_data.columns=['Age Ranges','Average Purchase Price']
```

```python
mean_data
```

```python
#Purchase Count
count_data=purchase_data.groupby('Age Ranges').count()
count_data=count_data.iloc[:, 1:2].reset_index()
count_data.columns=['Age Ranges', 'Purchase Count']

count_data
```

```python
#Total Purchase Value
sum_data=purchase_data.groupby('Age Ranges').sum()
sum_data=sum_data.iloc[:, 3:4].reset_index()
sum_data.columns=['Age Ranges','Total Purchase Value']
```

```python
sum_data
```

```python
#merging the above
final_df2=count_data.merge(mean_data)
final_df2
```

```python
#merging again
final_df2=final_df2.merge(sum_data)
final_df2
```

```python
#getting unique SN
final_df2=final_df2.merge(unique_SN)
final_df2
```

```python
#calculating the Average Purchase Total per Person by Age Group

final_df2['Avg Total Purchase per Person']= final_df2['Total Purchase Value']/final_df2['Unique people']
final_df2
```

```python
#deleting unique people
final_df2=final_df2.drop(columns=['Unique people'])
```

```python
#rounding
final_df2=final_df2.round(2)
```

```python
#set index
final_df2=final_df2.set_index('Age Ranges')
#final dataframe
final_df2
```

```python
#TOP SPENDERS
#reordering the columns  by renaming it first
top_spenders=purchase_data.groupby('SN').sum().iloc[:,3:4]

top_spenders.columns=['Total Purchase Value2']
top_spenders = top_spenders.reset_index()

top_spenders
```

```python
count=purchase_data.groupby('SN').count().sort_values(by=['Price']).iloc[:, :1]

count.columns=['Purchase Count']
count=count.reset_index()
count
```

```python
final_df3=top_spenders.merge(count)
final_df3
```

```python
final_df3['Average Purchase Price']=final_df3['Total Purchase Value2']/final_df3['Purchase Count']
```

```python
final_df3=final_df3.sort_values(by='Total Purchase Value2', ascending=False).head(5)
final_df3
```

```python
final_df3['Total Purchase Value']=final_df3['Total Purchase Value2']
final_df3
```

```python
final_df3=final_df3.drop(columns=['Total Purchase Value2'])
final_df3
```

```python
final_df3=final_df3.set_index('SN')
final_df3
```

```python
final_df3=final_df3.round(2)
final_df3
```
